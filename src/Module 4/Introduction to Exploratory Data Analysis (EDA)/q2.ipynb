**Title**: Handling Outliers<br>

**Task**: Detect and propose handling methods for outliers in the numerical columns of the datasets.<br>

Task 1: Age in Employee Dataset<br>
Task 2: Price in Product Sales Dataset<br>
Task 3: Math Score in Student Grades Dataset<br>

Instructions:<br>

Use box plots to visualize potential outliers.<br>
Suggest methods to handle them, such as removal or transformation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Sample DataFrames (using potentially more varied data for outlier demonstration) ---
employee_data_outliers = {
    'Employee ID': range(101, 116),
    'Name': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O'],
    'Age': [25, 30, 28, 35, 22, 40, 27, 31, 29, 33, 65, 26, 32, 24, 70],
    'Department': ['Sales'] * 5 + ['Marketing'] * 5 + ['Engineering'] * 5,
    'Salary': [50000, 60000, 55000, 70000, 48000, 62000, 58000, 72000, 52000, 65000, 80000, 53000, 68000, 51000, 75000],
    'Joining Date': pd.to_datetime(['2023-01-15', '2022-05-20', '2023-03-10', '2021-11-01', '2024-02-25',
                                      '2023-07-01', '2022-12-10', '2024-04-05', '2023-09-18', '2022-08-22',
                                      '2021-06-01', '2023-11-30', '2024-01-20', '2022-03-15', '2021-09-01'])
}
employee_df_outliers = pd.DataFrame(employee_data_outliers)

sales_data_outliers = {
    'Product ID': range(1, 16),
    'Product Name': [f'Product_{i}' for i in range(1, 16)],
    'Price': [100, 120, 110, 130, 115, 150, 160, 140, 125, 135, 50, 90, 80, 70, 500],
    'Quantity Sold': [10, 15, 12, 18, 20, 8, 7, 9, 11, 13, 25, 30, 35, 40, 5],
    'Category': ['A'] * 5 + ['B'] * 5 + ['C'] * 5,
    'Sales Date': pd.to_datetime(['2024-05-01'] * 15)
}
sales_df_outliers = pd.DataFrame(sales_data_outliers)

grades_data_outliers = {
    'Student ID': range(1, 16),
    'Student Name': [f'Student_{i}' for i in range(1, 16)],
    'Math Score': [85, 92, 78, 95, 88, 70, 65, 72, 80, 83, 98, 99, 100, 60, 40],
    'Science Score': [90, 88, 82, 91, 86, 75, 70, 78, 85, 89, 95, 92, 98, 65, 55],
    'English Score': [76, 80, 85, 79, 90, 68, 62, 70, 78, 82, 92, 95, 90, 58, 50],
    'Year': [2023] * 15
}
grades_df_outliers = pd.DataFrame(grades_data_outliers)

# --- Task 1: Age in Employee Dataset ---
print("--- Task 1: Age in Employee Dataset ---")
plt.figure(figsize=(8, 6))
sns.boxplot(y=employee_df_outliers['Age'])
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()

print("\nPotential Outliers in Age:")
Q1_age = employee_df_outliers['Age'].quantile(0.25)
Q3_age = employee_df_outliers['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age
outliers_age = employee_df_outliers[(employee_df_outliers['Age'] < lower_bound_age) | (employee_df_outliers['Age'] > upper_bound_age)]
print(outliers_age['Age'])

print("\nProposed Handling Methods for Age Outliers:")
print("- **Removal:** If these ages represent genuine but rare cases and the dataset is large enough, these rows could be removed. However, be cautious as extreme but valid data points might be lost.")
print("- **Capping/Flooring:** The outliers could be capped at the upper whisker value and floored at the lower whisker value. This retains the data points but reduces the impact of the extreme values.")
print("- **Transformation:** A transformation like log transformation might help if the age distribution is skewed, but it might not directly address extreme outliers. Winsorizing could also be used to limit extreme values to a certain percentile.")
print("-" * 50)

# --- Task 2: Price in Product Sales Dataset ---
print("\n--- Task 2: Price in Product Sales Dataset ---")
plt.figure(figsize=(8, 6))
sns.boxplot(y=sales_df_outliers['Price'])
plt.title('Box Plot of Price')
plt.ylabel('Price')
plt.show()

print("\nPotential Outliers in Price:")
Q1_price = sales_df_outliers['Price'].quantile(0.25)
Q3_price = sales_df_outliers['Price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_bound_price = Q1_price - 1.5 * IQR_price
upper_bound_price = Q3_price + 1.5 * IQR_price
outliers_price = sales_df_outliers[(sales_df_outliers['Price'] < lower_bound_price) | (sales_df_outliers['Price'] > upper_bound_price)]
print(outliers_price['Price'])

print("\nProposed Handling Methods for Price Outliers:")
print("- **Investigation:** It's crucial to investigate the reason for the high-priced product. It might be a valid entry for a premium product.")
print("- **Removal:** If the outlier is due to an error or is irrelevant to the analysis, it could be removed.")
print("- **Capping:** If the high price is valid but its extreme value disproportionately affects the analysis, capping it at a reasonable upper limit might be considered.")
print("- **Transformation:** Log transformation could help reduce the scale difference if there's a wide range of prices.")
print("-" * 50)

# --- Task 3: Math Score in Student Grades Dataset ---
print("\n--- Task 3: Math Score in Student Grades Dataset ---")
plt.figure(figsize=(8, 6))
sns.boxplot(y=grades_df_outliers['Math Score'])
plt.title('Box Plot of Math Score')
plt.ylabel('Math Score')
plt.show()

print("\nPotential Outliers in Math Score:")
Q1_math = grades_df_outliers['Math Score'].quantile(0.25)
Q3_math = grades_df_outliers['Math Score'].quantile(0.75)
IQR_math = Q3_math - Q1_math
lower_bound_math = Q1_math - 1.5 * IQR_math
upper_bound_math = Q3_math + 1.5 * IQR_math
outliers_math = grades_df_outliers[(grades_df_outliers['Math Score'] < lower_bound_math) | (grades_df_outliers['Math Score'] > upper_bound_math)]
print(outliers_math['Math Score'])

print("\nProposed Handling Methods for Math Score Outliers:")
print("- **Investigation:** Scores significantly lower than the rest might indicate students needing extra support or potential data entry errors. Very high scores might be genuine talent or, again, errors.")
print("- **Capping/Flooring:** If the outliers are likely errors, capping the lower scores at a minimum reasonable score (e.g., 0 if the scale starts there) and flooring the higher scores at a maximum (e.g., 100 if that's the max) could be an option.")
print("- **Removal (with caution):** Removing students based on scores should be done very carefully and with a clear understanding of the implications (e.g., if the analysis aims to study the performance of the entire student body).")
print("- **Transformation:** Transformations are less likely to be directly helpful for bounded scores like grades.")
print("-" * 50)
```

**Title**: Finding Relationships Between Features<br>

**Task**: Identify relationships between pairs of features in the datasets.<br>

Task 2: Price vs Quantity Sold in Product Sales Dataset<br>
Task 3: Math Score vs Science Score in Student Grades Dataset

Instructions: <br>

Use scatter plots or correlation coefficients to analyze the relationships.<br>
Describe any insights or patterns observed.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Sample DataFrames (using the outlier datasets from previous tasks) ---
employee_data_relation = {
    'Age': [25, 30, 28, 35, 22, 40, 27, 31, 29, 33, 65, 26, 32, 24, 70],
    'Salary': [50000, 60000, 55000, 70000, 48000, 62000, 58000, 72000, 52000, 65000, 80000, 53000, 68000, 51000, 75000]
}
employee_df_relation = pd.DataFrame(employee_data_relation)

sales_data_relation = {
    'Price': [100, 120, 110, 130, 115, 150, 160, 140, 125, 135, 50, 90, 80, 70, 500],
    'Quantity Sold': [10, 15, 12, 18, 20, 8, 7, 9, 11, 13, 25, 30, 35, 40, 5]
}
sales_df_relation = pd.DataFrame(sales_data_relation)

grades_data_relation = {
    'Math Score': [85, 92, 78, 95, 88, 70, 65, 72, 80, 83, 98, 99, 100, 60, 40],
    'Science Score': [90, 88, 82, 91, 86, 75, 70, 78, 85, 89, 95, 92, 98, 65, 55]
}
grades_df_relation = pd.DataFrame(grades_data_relation)

# --- Task 1: Salary vs Age in Employee Dataset ---
print("--- Task 1: Salary vs Age ---")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Salary', data=employee_df_relation)
plt.title('Scatter Plot of Salary vs Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

correlation_age_salary = employee_df_relation['Age'].corr(employee_df_relation['Salary'])
print(f"\nCorrelation between Age and Salary: {correlation_age_salary:.2f}")
print("\nInsights:")
print("- There appears to be a weak positive correlation between age and salary. Generally, salary tends to increase with age, but the relationship isn't very strong or linear based on this sample.")
print("- There might be other factors (like experience, role, department) that influence salary more significantly than just age.")
print("-" * 50)

# --- Task 2: Price vs Quantity Sold in Product Sales Dataset ---
print("\n--- Task 2: Price vs Quantity Sold ---")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price', y='Quantity Sold', data=sales_df_relation)
plt.title('Scatter Plot of Price vs Quantity Sold')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.show()

correlation_price_quantity = sales_df_relation['Price'].corr(sales_df_relation['Quantity Sold'])
print(f"\nCorrelation between Price and Quantity Sold: {correlation_price_quantity:.2f}")
print("\nInsights:")
print("- There seems to be a weak negative correlation between the price of a product and the quantity sold. Higher-priced items tend to have lower quantities sold, and vice versa.")
print("- The presence of a very high-priced item with a low quantity sold (potential outlier) could be influencing this correlation. Further investigation might be needed.")
print("-" * 50)

# --- Task 3: Math Score vs Science Score in Student Grades Dataset ---
print("\n--- Task 3: Math Score vs Science Score ---")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Math Score', y='Science Score', data=grades_df_relation)
plt.title('Scatter Plot of Math Score vs Science Score')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.show()

correlation_math_science = grades_df_relation['Math Score'].corr(grades_df_relation['Science Score'])
print(f"\nCorrelation between Math Score and Science Score: {correlation_math_science:.2f}")
print("\nInsights:")
print("- There is a strong positive correlation between Math Score and Science Score. Students who perform well in Math tend to perform well in Science, and vice versa.")
print("- This suggests that there might be some underlying common factors (like general academic ability or study habits) that influence performance in both subjects.")
print("-" * 50)
```
