```python
# Question 2: Geometry of PCA
# Description: Compute PCA on a simple 2D matrix and manually derive its principal components.

import numpy as np
import matplotlib.pyplot as plt

X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0],
              [2.3, 2.7],
              [2, 1.6],
              [1, 1.1],
              [1.5, 1.6],
              [1.1, 0.9]])

X_meaned = X - np.mean(X, axis=0)

cov_matrix = np.cov(X_meaned, rowvar=False)

eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:, sorted_idx]

X_pca = X_meaned @ eigenvectors[:, :2]

plt.figure(figsize=(8, 6))
plt.scatter(X_meaned[:, 0], X_meaned[:, 1], alpha=0.7)
origin = np.mean(X, axis=0)
for i in range(2):
    vec = eigenvectors[:, i] * eigenvalues[i]
    plt.quiver(*origin, *vec, angles='xy', scale_units='xy', scale=1, color=['r', 'g'][i])
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.title('PCA Geometry on 2D Data')
plt.axis('equal')
plt.grid(True)
plt.show()
```
