```python
# Question 5: Visual Comparison of PCA and t-SNE
# Description: Compare PCA and t-SNE on the same dataset visually.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Step 1: Load and normalize the dataset
digits = load_digits()
X = digits.data
y = digits.target

# Standardize the data (important for PCA and t-SNE)
X_scaled = StandardScaler().fit_transform(X)

# Step 2: Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 3: Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Step 4: Plot the results side by side
plt.figure(figsize=(14, 6))

# PCA Plot
plt.subplot(1, 2, 1)
scatter1 = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', s=40, alpha=0.7)
plt.title("PCA - Digits Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(scatter1, ticks=range(10))

# t-SNE Plot
plt.subplot(1, 2, 2)
scatter2 = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=40, alpha=0.7)
plt.title("t-SNE - Digits Dataset")
plt.xlabel("t-SNE Feature 1")
plt.ylabel("t-SNE Feature 2")
plt.colorbar(scatter2, ticks=range(10))

plt.tight_layout()
plt.show()
```
