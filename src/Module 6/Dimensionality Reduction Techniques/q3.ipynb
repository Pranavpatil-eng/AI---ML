```python
# Question 3: Implement t-SNE on a Dataset
# Description: Use t-SNE to visualize high-dimensional data.

# Import required libraries
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Step 1: Load the dataset
digits = load_digits()
X = digits.data     # High-dimensional data (64 features)
y = digits.target   # Labels (digits 0-9)

# Step 2: Normalize the data
X_scaled = StandardScaler().fit_transform(X)

# Step 3: Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_embedded = tsne.fit_transform(X_scaled)

# Step 4: Plot the result
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='tab10', s=50, alpha=0.7)
plt.colorbar(scatter, ticks=range(10))
plt.title("t-SNE Visualization of Digits Dataset")
plt.xlabel("t-SNE Feature 1")
plt.ylabel("t-SNE Feature 2")
plt.grid(True)
plt.show()
```
