```python
# Question 8: Clustering Text Data
# Description: Apply K-Means clustering to a set of text documents (e.g., news articles) after using TF-IDF for feature extraction.

# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Example text documents (you can replace this with your dataset)
documents = [
    "The economy is growing and jobs are being created.",
    "The stock market has had a significant drop in recent months.",
    "Government policies are influencing economic growth.",
    "The climate change crisis is urgent and requires immediate action.",
    "Global warming is a serious environmental issue that needs attention.",
    "Environmental protection laws are becoming stricter in many countries.",
    "Sports events are important to many communities.",
    "Football is one of the most popular sports worldwide.",
    "The sports industry has been growing in recent years.",
    "Technology is transforming the way we live and work.",
    "Artificial intelligence is advancing rapidly in various industries.",
    "The tech industry is booming with innovations in AI, robotics, and software."
]

# Step 1: Feature extraction using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Step 2: Apply KMeans clustering
n_clusters = 3  # Define the number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Step 3: Assign the documents to clusters
labels = kmeans.labels_

# Step 4: Print out the cluster assignments
for i, doc in enumerate(documents):
    print(f"Document {i+1}: Cluster {labels[i]}")

# Step 5: Optionally, visualize the clusters (using PCA for dimensionality reduction)
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X.toarray())

plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='viridis', s=100)
plt.title("K-Means Clustering of Text Documents (PCA reduced)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar()
plt.show()
```
