```python
# Question 6 : Comparing K-Means and Hierarchical Clustering
# Description: Compare the results of K-Means and Hierarchical Clustering on a dataset and discuss the potential differences in their clustering patterns.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Generate a synthetic dataset (You can replace this with your own dataset)
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans_labels = kmeans.fit_predict(X)

# Step 3: Apply Hierarchical Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_labels = agg_clustering.fit_predict(X)

# Step 4: Visualize K-Means clustering results
plt.figure(figsize=(12, 6))

# Subplot 1: K-Means clustering result
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='o', s=50)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Subplot 2: Hierarchical clustering result
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=agg_labels, cmap='viridis', marker='o', s=50)
plt.title("Hierarchical Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()

# Step 5: Visualize the dendrogram for Hierarchical Clustering
plt.figure(figsize=(8, 6))
Z = linkage(X, 'ward')  # 'ward' minimizes the variance of merged clusters
dendrogram(Z)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```
