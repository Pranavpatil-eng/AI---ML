```python
# Question 8: Parameter Tuning in DBSCAN
# Description: Experiment with different values of eps and min_samples in DBSCAN and observe the effect on clustering results.

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

# Step 1: Generate synthetic data (You can replace this with your own dataset)
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Step 2: Function to apply DBSCAN and visualize the results
def apply_dbscan(eps, min_samples):
    # Apply DBSCAN with specified parameters
    db = DBSCAN(eps=eps, min_samples=min_samples)
    y_db = db.fit_predict(X)

    # Plot the clustering results
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y_db, cmap='viridis', marker='o', s=50)
    plt.title(f"DBSCAN Clustering (eps={eps}, min_samples={min_samples})")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.colorbar()
    plt.show()

# Step 3: Experiment with different values of eps and min_samples
# Experiment 1: Small eps, large min_samples
apply_dbscan(eps=0.2, min_samples=5)

# Experiment 2: Larger eps, smaller min_samples
apply_dbscan(eps=1.0, min_samples=5)

# Experiment 3: Large eps, small min_samples
apply_dbscan(eps=1.5, min_samples=3)

# Experiment 4: Small eps, large min_samples
apply_dbscan(eps=0.4, min_samples=10)

# Experiment 5: Larger eps, large min_samples
apply_dbscan(eps=1.0, min_samples=10)
```
